```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
print(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
print(y)
```

```python
print(np.unique(y))
```

```python
plt.figure(figsize=(7, 5))
for target in np.unique(y):
    plt.scatter(X_pca[y == target, 0], X_pca[y == target, 1], label=iris.target_names[target])
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.title("Iris Dataset: PCA (2D Projection)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(7, 5))
for target in np.unique(y):
    plt.scatter(X[y == target, 0], X[y == target, 1],  label=iris.target_names[target])
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.title("Iris Dataset: PCA (2D Projection)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total retained variance:", np.sum(pca.explained_variance_ratio_))
```
